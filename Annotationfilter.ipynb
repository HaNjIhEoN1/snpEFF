```python
def genotypeMatrix(line):
    """
    Replace with Genotype information
    """
    
    allele = np.array([line[3]] + line[4].split(','))
    
    def change_gt(i):
        try:
            ix = list(map(int, i.split(':')[0].split('/')))
            gt = '/'.join(allele[ix])
        except ValueError:
            gt = 'N/N'
        return gt
    
    return [change_gt(i) for i in line[9:]]
```
